```python
import numpy as np
import cv2 as cv
import json
```

```python
def load_annotations(image_path: str):
    annotation_path = image_path.replace(".png", ".coco.json")
    with open(annotation_path, "r") as f:
        annotations = json.load(f)
    return annotations
```

```python
def get_image_and_seg_points(image_path: str):
    image = cv.imread(image_path)
    annotations = load_annotations(image_path)
    return image, np.reshape(np.array(annotations['annotations'][0]["segmentation"], dtype=np.int32), (-1, 2))
```

```python
root_image_dir = "/home/maco/Documents/photoneo/data/groceries/train"
image_path =  root_image_dir + "/mixed/billa/20210223_groceries_0032.png"
```

```python
img, seg_points = get_image_and_seg_points(image_path)
```

```python
seg_points.shape
```

```python
cv.imshow('Original image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import torchvision.transforms as transforms
tensor_img = transforms.ToTensor()(img)
resize_transform = transforms.Resize(size=(640, 640))
resized_image = resize_transform(tensor_img)
```

```python
test_img = resized_image.permute(1, 2, 0).numpy()
cv.imshow('Original image', test_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Crop segmentation region

```python
mask = np.zeros_like(img)
```

```python
mask.shape
```

```python
cv.fillPoly(mask, [seg_points], (255, 255, 255))
```

```python
cv.imshow('Segmented Image', mask)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_copy = img.copy()
```

```python
segmented_image = cv.bitwise_and(img_copy, mask)
cv.imshow('Segmented Image', segmented_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Color segmentation region

```python
img_copy2 = img.copy()
```

```python
img_copy2
```

```python
masked_img = np.where(mask, [0,255,0], img_copy2)
```

```python
masked_img = masked_img.astype(np.uint8)
```

```python
out = cv.addWeighted(img_copy2, 0.8, masked_img, 0.2,0)
```

```python
cv.imshow('Segmented Image', out)
cv.waitKey(0)
cv.destroyAllWindows()
```
